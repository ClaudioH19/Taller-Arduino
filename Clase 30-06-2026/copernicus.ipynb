```python
import cdsapi

############################
#Definir latitud y longitud#
############################
monthly = True #<--- Cambiar a True si se quiere descargar todo el mes de junio 2023

#lat= -35.404409
#lon= -71.633446

# Para Licantén, cambiar los valores comentados
# Si usan Licantén esta opción cambiar la variable monthly a monthly=True
lat = -34.98461
lon = -71.98549



dias=[]
for i in range(1, 31):
    dias.append(str(i).zfill(2))



dataset = "reanalysis-era5-land"
request = {
    "variable": [
        "2m_dewpoint_temperature",
        "2m_temperature",
        "surface_solar_radiation_downwards",
        "10m_u_component_of_wind",
        "10m_v_component_of_wind",
        "surface_pressure",
        "total_precipitation"
    ],
    "year": "2023" if monthly else "2026",
    "month": "06",
    "day": dias if monthly else ["09"],
    "time": [
        "00:00", "01:00", "02:00",
        "03:00", "04:00", "05:00",
        "06:00", "07:00", "08:00",
        "09:00", "10:00", "11:00",
        "12:00", "13:00", "14:00",
        "15:00", "16:00", "17:00",
        "18:00", "19:00", "20:00",
        "21:00", "22:00", "23:00"
    ],
    "data_format": "netcdf",
    "download_format": "unarchived",
    "area": [lat+0.5, lon-0.5, lat-0.5, lon+0.5]
}

client = cdsapi.Client()
client.retrieve(dataset, request).download()
```

```python
import math
import xarray as xr
import os 
#encontrar el archivo .nc descargado
for file in os.listdir("./"):
    if file.endswith(".nc"):
        ruta_nc = file
        break

ds = xr.open_dataset(ruta_nc,engine="netcdf4")

# Truncar a 1 decimales
lat_objetivo = math.trunc(lat * 10) / 10
lon_objetivo = math.trunc(lon * 10) / 10


df = ds.to_dataframe()
df = df.reset_index()

# Filtrar solo la celda que coincide con la lat/lon truncada
df = df[
    (df["latitude"].round(1) == lat_objetivo) &
    (df["longitude"].round(1) == lon_objetivo)
]

#calcular humedad relativa
def exp(t: float):
    return math.e ** ((17.625 * t) / (243.04 + t))

def es(t: float):
    return 6.112 * exp(t)

#Humedad relativa
df["humidity"] = ((es(df["d2m"]) / es(df["t2m"])) * 100)

# definir nombre
nombre_salida = f"data_{lat_objetivo}_{lon_objetivo}.csv"
df.to_csv(nombre_salida, index=False)
print("Hecho")
```
